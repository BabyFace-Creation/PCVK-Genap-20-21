```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import math
from PIL import Image as im
```

```python
image = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/mandrill.tiff')
img_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])
image_sharp = cv.filter2D(src=img_gray, ddepth=-1, kernel=kernel)
cv2_imshow(image_sharp)
```

```python
image = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/mandrill.tiff')
img_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

sigma=math.sqrt(441)
gaussian_kernel = cv.getGaussianKernel(21, sigma)
gauss_kernel = gaussian_kernel 
gaussian_kernel.transpose()
image_sharp = cv.filter2D(src=img_gray, ddepth=-1, kernel=gauss_kernel)

cv2_imshow(image_sharp)
```
